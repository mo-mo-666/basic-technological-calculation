# Discrete Fourier transform
Discrete Fourier transform
$$A_k = \sum_{m=0}^{N-1} a_m \exp\left\{-2\pi i{mk \over N}\right\} \qquad k = 0,\ldots,N-1.$$

Inverse discrete Fourier transform
$$a_m = \frac{1}{n}\sum_{k=0}^{N-1}A_k\exp\left\{2\pi i{mk\over N}\right\} \qquad m = 0,\ldots,N-1.$$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
```

# Fourier transform
$$F[f](\xi)$$

$$F[1_{(-1, 1)}](\xi) = \begin{cases}
\frac{e^{-ib\xi}-e^{-ia\xi}}{-i\xi} & \xi \neq 0\\
b-a & \xi = 0
\end{cases}
$$

```python
a = -2
b= 2
datanum = 100
step = (b-a) / datanum

x = np.linspace(a, b, datanum)
f = [0] * datanum + [1] * datanum + [0] * datanum
F = step*np.fft.fft(f)
plt.plot(np.abs(F))
plt.show()
print(np.abs(F))
```
